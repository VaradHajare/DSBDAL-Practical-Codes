```python
'''Text Analytics 
1. Extract Sample document and apply following document preprocessing methods: 
Tokenization, POS Tagging, stop words removal, Stemming and Lemmatization. 
2. Create representation of document by calculating Term Frequency and Inverse Document 
Frequency.'''
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
import re
import pandas as pd
import numpy as np
```

```python
def preprocess_text(text, include_pos_stem=False):
    cleaned_doc = re.sub(r'[^a-zA-Z\s]', '', text).lower()
    initial_tokens = word_tokenize(cleaned_doc)

    stop_words = set(stopwords.words('english'))
    filtered_tokens = [word for word in initial_tokens if word not in stop_words]

    lemmatizer = WordNetLemmatizer()
    lemmatized_tokens = [lemmatizer.lemmatize(word) for word in filtered_tokens]

    if include_pos_stem:
        pos_tags = pos_tag(initial_tokens) # POS tagging on initial tokens
        stemmer = PorterStemmer()
        stemmed_tokens = [stemmer.stem(word) for word in filtered_tokens]
        return initial_tokens, pos_tags, filtered_tokens, stemmed_tokens, lemmatized_tokens
    else:
        return lemmatized_tokens
```

```python
# List of NLTK packages to download
nltk_data_packages = [
    'punkt',
    'punkt_tab', 
    'averaged_perceptron_tagger',
    'averaged_perceptron_tagger_eng', 
    'stopwords',
    'wordnet'
]

for package in nltk_data_packages:
    nltk.download(package, quiet=True, raise_on_error=False)
```

```python
sample_document = """Text analytics is the process of discovering patterns and insights from unstructured text data. It involves various techniques from natural language processing, machine learning, and statistics to transform raw text into meaningful and structured information. This can be used for sentiment analysis, topic modeling, named entity recognition, and more. For example, businesses use text analytics to understand customer feedback, monitor social media, and improve product development. It is a powerful tool for data-driven decision making."""

print("Original Document:")
print(sample_document)
```

```python
# Tokenization
# Tokenization is the process of breaking down a text into smaller units, such as words or phrases, called tokens.
# Remove special characters and digits, convert to lowercase
cleaned_document = re.sub(r'[^a-zA-Z\s]', '', sample_document).lower()

tokens = word_tokenize(cleaned_document)

print("\nTokens:")
print(tokens)
```

```python
# POS Tagging
# Part-of-Speech (POS) tagging assigns a grammatical category (e.g., noun, verb, adjective) to each token.

pos_tags = pos_tag(tokens)

print("\nPOS Tags:")
print(pos_tags)
```

```python
# Stop Words Removal
# Stop words are common words (e.g., 'the', 'is', 'and') that often carry little meaning and can be removed to reduce noise.

stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word not in stop_words]

print("\nTokens after Stop Words Removal:")
print(filtered_tokens)
```

```python
# Stemming
# Stemming reduces words to their root form (stem) by removing suffixes, even if the root is not a valid word.

stemmer = PorterStemmer()
stemmed_tokens = [stemmer.stem(word) for word in filtered_tokens]

print("\nStemmed Tokens:")
print(stemmed_tokens)
```

```python
# Lemmatization
# Lemmatization reduces words to their base or dictionary form (lemma), ensuring the result is a valid word.

lemmatizer = WordNetLemmatizer()
lemmatized_tokens = [lemmatizer.lemmatize(word) for word in filtered_tokens]

print("\nLemmatized Tokens:")
print(lemmatized_tokens)
```

```python
def calculate_tf_idf(documents):
    # Tokenize and preprocess each document
    processed_documents = []
    all_words = set()
    for doc in documents:
        cleaned_doc = re.sub(r'[^a-zA-Z\s]', '', doc).lower()
        tokens = word_tokenize(cleaned_doc)
        filtered_tokens = [word for word in tokens if word not in stop_words]
        lemmatized_tokens = [lemmatizer.lemmatize(word) for word in filtered_tokens]
        processed_documents.append(lemmatized_tokens)
        all_words.update(lemmatized_tokens)

    # Sort words for consistent order
    all_words = sorted(list(all_words))

    # Calculate Term Frequency (TF)
    tf_matrix = np.zeros((len(documents), len(all_words)))
    for i, doc_tokens in enumerate(processed_documents):
        for word in doc_tokens:
            if word in all_words:
                tf_matrix[i, all_words.index(word)] += 1

    # Normalize TF (word count / total words in document)
    for i in range(len(documents)):
        total_words_in_doc = sum(tf_matrix[i])
        if total_words_in_doc > 0:
            tf_matrix[i] = tf_matrix[i] / total_words_in_doc

    # Calculate Inverse Document Frequency (IDF)
    idf_vector = np.zeros(len(all_words))
    num_documents = len(documents)
    for i, word in enumerate(all_words):
        doc_count = sum(1 for doc_tokens in processed_documents if word in doc_tokens)
        idf_vector[i] = np.log(num_documents / (doc_count + 1)) # Add 1 to avoid division by zero

    # Calculate TF-IDF
    tf_idf_matrix = tf_matrix * idf_vector

    # Create a DataFrame for better visualization
    tf_idf_df = pd.DataFrame(tf_idf_matrix, columns=all_words)
    tf_df = pd.DataFrame(tf_matrix, columns=all_words)
    idf_df = pd.DataFrame([idf_vector], columns=all_words)

    return tf_df, idf_df, tf_idf_df

# Let's use a list of documents for TF-IDF calculation
documents = [
    "Text analytics is the process of discovering patterns and insights from unstructured text data.",
    "It involves various techniques from natural language processing, machine learning, and statistics to transform raw text into meaningful and structured information.",
    "This can be used for sentiment analysis, topic modeling, named entity recognition, and more. For example, businesses use text analytics to understand customer feedback, monitor social media, and improve product development.",
    "It is a powerful tool for data-driven decision making."
]

tf_df, idf_df, tf_idf_df = calculate_tf_idf(documents)

print("\nTerm Frequency (TF) Matrix:")
display(tf_df)

print("\nInverse Document Frequency (IDF) Vector:")
display(idf_df)

print("\nTF-IDF Matrix:")
display(tf_idf_df)
```
