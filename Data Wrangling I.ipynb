```python
# Step 1: Import Required Libraries
import pandas as pd
import numpy as np
```

```python
# Step 2: Locate and Load Dataset
# Dataset: Titanic Dataset
# Source: https://www.kaggle.com/datasets/yasserh/titanic-dataset

df = pd.read_csv("data/Titanic-Dataset.csv")

# Display first 5 rows
df.head()
```

```python
# Step 3: Check Dataset Dimensions
# Shows number of rows and columns

df.shape
```

```python
# Step 4: Dataset Information
# Displays column names, non-null counts, and data types

df.info()
```

```python
# Step 5: Statistical Summary
# Provides mean, std, min, max, quartiles for numerical columns

df.describe()
```

```python
# Step 6: Check Missing Values
# Counts missing (null) values in each column

df.isnull().sum()
```

```python
# Step 7: Handle Missing Values
# Age → fill with mean (numerical data)
# Embarked → fill with mode (categorical data)
# Drop Cabin column due to excessive missing values

df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df.drop(columns=['Cabin'], inplace=True)
```

```python
# Step 8: Verify Missing Values

df.isnull().sum()
```

```python
# Step 9: Check Data Types of Columns

df.dtypes
```

```python
# Step 10: Convert Data Types
# Convert Survived and Pclass into categorical types

df['Survived'] = df['Survived'].astype('category')
df['Pclass'] = df['Pclass'].astype('category')
```

```python
# Step 11: Normalize Numerical Data
# Apply Min-Max Scaling to Fare column → values between 0 and 1

df['Fare_norm'] = (df['Fare'] - df['Fare'].min()) / (df['Fare'].max() - df['Fare'].min())

# View original vs normalized
df[['Fare', 'Fare_norm']].head()
```

```python
# Step 12: Convert Categorical Variables to Numerical

df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
```
