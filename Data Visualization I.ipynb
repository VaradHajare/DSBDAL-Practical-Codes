```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
titanic = sns.load_dataset("titanic")

print("First 5 Rows of the Dataset:")
print(titanic.head())
```

```python
print("\nShape of the Dataset:")
print(titanic.shape)
```

```python
print("\nDataset Information:")
print(titanic.info())
```

```python
# 1. Count plot to see survival distribution
plt.figure()
sns.countplot(x='survived', data=titanic)
plt.title("Survival Count (0 = Not Survived, 1 = Survived)")
plt.xlabel("Survival Status")
plt.ylabel("Number of Passengers")
plt.show()
```

```python
# 2. Survival based on Gender
plt.figure()
sns.countplot(x='sex', hue='survived', data=titanic)
plt.title("Survival Based on Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Passengers")
plt.show()
```

```python
# 3. Survival based on Passenger Class
plt.figure()
sns.countplot(x='class', hue='survived', data=titanic)
plt.title("Survival Based on Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Number of Passengers")
plt.show()
```

```python
# Distribution of Ticket Fare (Histogram)

plt.figure()

# Plot histogram of fare column
sns.histplot(titanic['fare'], bins=30, kde=True)

plt.title("Distribution of Ticket Fare")
plt.xlabel("Fare Amount")
plt.ylabel("Number of Passengers")

plt.show()
```
