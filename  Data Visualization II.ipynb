```python
'''Data Visualization II 
1. Use the inbuilt dataset 'titanic' as used in the above problem. Plot a box plot for distribution 
of age with respect to each gender along with the information about whether they survived 
or not. (Column names : 'sex' and 'age') 
2. Write observations on the inference from the above statistics.'''
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the titanic dataset
titanic_df = sns.load_dataset('titanic')

# Display the first few rows to understand the data structure
print("First 5 rows of the Titanic dataset:")
display(titanic_df.head())
```

```python
# Create a box plot for age distribution with respect to gender and survival status
plt.figure(figsize=(10, 6))
sns.boxplot(x='sex', y='age', hue='survived', data=titanic_df, palette='viridis')
plt.title('Age Distribution by Gender and Survival Status on the Titanic')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.xticks(ticks=[0, 1], labels=['Male', 'Female']) # Assuming 'male' and 'female' are default order
plt.legend(title='Survived', loc='upper right', labels=['No', 'Yes'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
